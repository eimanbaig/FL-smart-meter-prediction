```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# onlyfiles = [f for f in listdir('./num_clust_four_100/pred') if isfile(join('./no_clust_results/pred', f))and f[-4:]==".csv"]
# rm=[]
# for f in onlyfiles:
kmeans_df = pd.read_csv('./num_clust_four_100/pred-kmeans/1660.csv', index_col=0)
test_df = pd.read_csv('./num_clust_four_100/test-kmeans/1660.csv', index_col=0)
fuzzycmeans_df = pd.read_csv('./num_clust_four_100/pred-fuzzycmeans/1660.csv', index_col=0)
hierarchical_df = pd.read_csv('./num_clust_four_100/pred-hierarchical/1660.csv', index_col=0)
minibatchkmeans_df = pd.read_csv('./num_clust_four_100/pred-minibatchkmeans/1660.csv', index_col=0)
no_clust_df = pd.read_csv('./no_clustering_100/pred-no_clustering/1660.csv', index_col=0)

kmeans = [kmeans_df.iloc()[i].tolist() for i in range(len(kmeans_df))]
test = [test_df.iloc()[i].tolist() for i in range(len(test_df))]
fuzzycmeans = [fuzzycmeans_df.iloc()[i].tolist() for i in range(len(fuzzycmeans_df))]
hierarchical = [hierarchical_df.iloc()[i].tolist() for i in range(len(hierarchical_df))]
minibatchkmeans = [minibatchkmeans_df.iloc()[i].tolist() for i in range(len(minibatchkmeans_df))]
no_clust = [no_clust_df.iloc()[i].tolist() for i in range(len(no_clust_df))]

# print(pred[0], pred[1])

kmeans_pred=[]
fuzzycmeans_pred=[]
hierarchical_pred=[]
minibatchkmeans_pred=[]
actualValue=[]
NoClustering=[]
for i in range(len(kmeans)):
    sum1 = 0
    sum2 = 0
    sum3 = 0
    sum4 = 0
    sum5=0
    sum6=0
    for j in range(len(kmeans[i])):
        sum1+=kmeans[i][j]
        sum2+=fuzzycmeans[i][j]
        sum3+=hierarchical[i][j]
        sum4+=minibatchkmeans[i][j]
        sum5+=test[i][j]
        sum6+=no_clust[i][j]
#         print(sum1, sum2)
    kmeans_pred.append(sum1/len(kmeans[i]))
    fuzzycmeans_pred.append(sum2/len(fuzzycmeans[i]))
    hierarchical_pred.append(sum3/len(hierarchical[i]))
    minibatchkmeans_pred.append(sum4/len(minibatchkmeans[i]))
    actualValue.append(sum5/len(test[i]))
    NoClustering.append(sum6/len(no_clust[i]))

# kmeans_pred=[]
# fuzzycmeans_pred=[]
# hierarchical_pred=[]
# minibatchkmeans_pred=[]
# actualValue=[]
# for i in range(len(kmeans)):
#     for j in range(len(kmeans[i])):
#         kmeans_pred.append(kmeans[i][j])
#         fuzzycmeans_pred.append(fuzzycmeans[i][j])
#         hierarchical_pred.append(hierarchical[i][j])
#         minibatchkmeans_pred.append(minibatchkmeans[i][j])
#         actualValue.append(test[i][j])
```

```python
def KL(a, b):
    a = np.asarray(a, dtype=np.float)
    b = np.asarray(b, dtype=np.float)

    return np.sum(np.where(a != 0, a * np.log(a / b), 0))


values1 = [1.346112,1.337432,1.246655]
values2 = [1.033836,1.082015,1.117323]

print KL(values1, values2)
```
